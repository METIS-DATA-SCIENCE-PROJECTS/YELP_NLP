##2.DATA PREPROCESSING

```python
import string
string.punctuation
```

```python
df_review_new = df_review[['stars','text']]
df_review_new
```

```python
#Function to remove Punctuation
def remove_punct(text):
    text_nopunct = "".join([char for char in text if char not in string.punctuation])# It will discard all punctuations
    return text_nopunct

df_review_new['text_clean'] = df_review_new['text'].apply(lambda x: remove_punct(x))

df_review_new.head()
```

#Remove Numbers

```python
def clean_text(text):
    # remove numbers
    text_nonum = re.sub(r'\d+', '', text)
    return text_nonum

df_review_new['nonum'] = df_review_new['text_clean'].apply(lambda x: clean_text(x))

df_review_new.head()
```

#Tokenization

```python
import re

# Function to Tokenize words
def tokenize(text):
    tokens = re.split('\W+', text) #W+ means that either a word character (A-Za-z0-9_) or a dash (-) can go there.
    return tokens

df_review_new['text_tokenized'] = df_review_new['nonum'].apply(lambda x: tokenize(x.lower())) 
#We convert to lower as Python is case-sensitive. 

df_review_new.head()
```

#Remove Stop Words

```python
 import nltk 
 nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

stopword = stopwords.words('english')# All English Stopwords
```

```python
# Function to remove Stopwords
def remove_stopwords(tokenized_list):
    text = [word for word in tokenized_list if word not in stopword]# To remove all stopwords
    return text

df_review_new['text_nonstop'] = df_review_new['text_tokenized'].apply(lambda x: remove_stopwords(x))

df_review_new.head()
```

#Stemming

```python
from nltk.stem.lancaster import LancasterStemmer

stemmer = LancasterStemmer()

def stemming(tokenized_text):
    text = [stemmer.stem(word) for word in tokenized_text]
    return text

df_review_new['text_stemmed'] = df_review_new['text_nonstop'].apply(lambda x: stemming(x))

df_review_new.head()
```
